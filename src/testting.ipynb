```python
import optuna

def objective(trial):
    x = trial.suggest_uniform('x', -10, 10)
    score = (x - 2) ** 2
    print('x: %1.3f, score: %1.3f' % (x, score))
    return score

def objective2(trial):
    x = trial.suggest_uniform('x', -5, 5)
    score = (x - 3)**2 + 4
    print("x : %1.3f, score: %1.3f" % (x, score))
    return score

study = optuna.create_study()
study.optimize(objective, n_trials=100)

objective(optuna.trial.FixedTrial({'x': 1}))

print("best_valueを出したときのparamater")
print(study.best_params)

study_2 = optuna.create_study()
study_2.optimize(objective2, n_trials = 100)

print("best_valueを出したときのparamater")
print(study_2.best_params)

```
